```python
class BinarySearchTreeNode:
    
    def __init__(self,data):
        self.data=data
        self.left=None
        self.right=None
    
    def add(self,data):
        if data==self.data:
            return
        
        if data<self.data:
            if self.left:
                self.left.add(data)
            else:
                self.left=BinarySearchTreeNode(data)
        else:
            if self.right:
                self.right.add(data)
            else:
                self.right=BinarySearchTreeNode(data)
                
    def inorder(self):
        ele=[]
        
        if self.left:
            ele+=self.left.inorder()
            
        ele.append(self.data)
        
        if self.right:
            ele+=self.right.inorder()
        
        return ele
    
    def search(self,val):
        if self.data==val:
            return True
        
        if val<self.data:
            if self.left:
                return self.left.search(val)
            else:
                return False
        else:
            if self.right:
                return self.right.search(val)
            else:
                return False
            
    def find_min(self):
        if self.left is None:
            return self.data
        else:
            return self.left.find_min()
        
    def find_max(self):
        if self.right is None:
            return self.data
        else:
            return self.right.find_max()
        
    def calc_sum(self):
        arr=self.inorder()
        return sum(arr)
    def post(self):
        ele=[]
        
        if self.left:
            ele+=self.left.post()
        if self.right:
            ele+=self.right.post()
        ele.append(self.data)
        
        return ele
    def pre(self):
        ele=[]
        
        ele.append(self.data)
        if self.left:
            ele+=self.left.pre()
        if self.right:
            ele+=self.right.pre()
        
        return ele
    
    def delete(self,val):
        if val<self.data:
            if self.left:
                self.left= self.left.delete(val)
        elif val>self.data:
            if self.right:
                self.right=self.right.delete(val)
        else:
            if self.left is None and self.right is None:
                return None
            
            if self.left is None:
                return self.right
            elif self.right is None:
                return self.left
            
            m=self.left.find_max()
            self.data=m
            self.left=self.left.delete(m)
            
        return self
def build(ele):   
    root=BinarySearchTreeNode(ele[0])
    for i in range(1,len(ele)):
        root.add(ele[i])
    return root
    
            
if __name__=='__main__':
    data=[17,4,1,20,9,23,18,34]
    tree=build(data)
    print(tree.pre())
    print(tree.inorder())
    tree.delete(20)
    print(tree.inorder())
    print(tree.pre())
    
```
